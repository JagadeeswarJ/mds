17. NumPy Arrays - Arithmetic and Statistical Operations with Universal Functions

```python
import numpy as np

arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

print(arr1 + arr2)
print(arr1 * arr2)

print(np.mean(arr1))
print(np.std(arr1))
print(np.sum(arr1))
print(np.min(arr1))
print(np.max(arr1))
```
